# 2. Practical Application
Simulate a simple gradebook system:



```python
# Create a 2D array representing 5 students and their marks in 4 subjects
# rows=student_name
# column=sub_name
# sub_name=eng,nep,phy,Math
# student_name=s1,s2,s3,s4,s5
marks=np.array([[67,73,85,92],
               [54,97,52,69],
               [86,84,93,90],
               [95,89,92,93],
               [85,80,85,83]])
print(marks)
```

```python
# Calculate and print:
# Average marks for each student
avg_marks_student=np.mean(marks,axis=1)
print(avg_marks_student)
# Average marks for each subject
avg_marks_subjects=np.mean(marks,axis=0)
print(avg_marks_subjects)
# Identify which student has the highest total marks
total_marks=np.sum(marks,axis=1)
highest_student=np.argmax(total_marks)
students = ["S1", "S2", "S3", "S4", "S5"]
print("Highest marks scored by:", students[highest_student])
```
