# 1. Array Manipulation and Transformation



```python
import numpy as np
# Given the following 2D array:
data = np.array([[12, 5, 8, 15 ],
                 [3, 18, 9, 6 ],
                 [14, 7, 11, 20]])
```

```python
# Perform the following operations:

# a) Transpose the array and print the result
trans=data.transpose()
print(trans)
```

```python
# b) Flatten the array using both ravel() and flatten(), then modify the first element of each result. Explain the difference in behavior.
a=data.ravel()
b=data.flatten()
print(a)
print(b)
a[0]=2
print(data)
b[0]=3
print(data)
```

In above problem,ravel returns a view but flattens creates a new copy and data change is not possible in original data for flatten.

```python
# c) Reshape the array to shape (6, 2)
x=data.reshape(6,2)
print(x)
```

```python
# d) Create two separate arrays by splitting the original array vertically
np.hsplit(data,2)
```

```python
# e) Use np.where() to replace all values less than 10 with 0, and all other values with their square
np.where(data < 10, 0, data**2)
```

```python
# f) Create a masked array that masks all values greater than 15
import numpy.ma as ma
masked = ma.masked_where(data>15,data)   # mask negatives
print(masked)
```

```python
# g) Concatenate the original array with its transpose (hint: you'll need to reshape first to match dimensions)
trans_data=trans.reshape(3,4)
print(trans_data)
```

```python
concat=np.concatenate((data,trans_data),axis=0)
print(concat)
```

# 2. Image Filter Application using NumPy Arrays
Load an image (you can use any image file or create a sample image array) and apply any filters (eg., Grayscale, blur, negative, brightness) using NumPy operations.

```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

image_path = '/content/dog (1).png'
image = imread(image_path)

# Display original image
plt.imshow(image)
plt.title("Original Image")
plt.show()

# Grayscale filter
gray = np.mean(image, axis=2).astype(np.uint8)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```
